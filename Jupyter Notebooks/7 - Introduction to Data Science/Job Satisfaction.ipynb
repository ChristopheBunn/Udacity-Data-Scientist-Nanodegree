```python
import pandas as pd
import numpy as np
import JobSatisfaction as t
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('./survey_results_public.csv')
schema = pd.read_csv('./survey_results_schema.csv')
df.head()
```

```python
set(df.columns)
```

```python
# The proportion of missing values in the Job Satisfaction column
print(df.JobSatisfaction.head())
nulls = df.JobSatisfaction.isnull()
print(nulls.head())
num_nulls = 0
for b in nulls:
    if b:
        num_nulls += 1
print("The proportion of missing values in JobSatisfaction is ", num_nulls / df.shape[0])
print("This is a better way to calculate the proportion: ", df.JobSatisfaction.isnull().mean())
```

```python
# According to EmploymentStatus, which group has the highest average job satisfaction?
df[['EmploymentStatus', 'JobSatisfaction']][:10]
```

```python
df.groupby(['EmploymentStatus']).mean() # average of all numeric columns for each value of EmploymentStatus
```

```python
df.groupby(['EmploymentStatus']).mean()['JobSatisfaction'] # averaqe JobSatisfaction for each EmploymentStatus
```

```python
# In general, do smaller companies appear to have employees with higher job satisfaction?
df.groupby(['CompanySize']).mean()['JobSatisfaction'].sort_values(ascending=False) # avg JS for each CompanySize
```
